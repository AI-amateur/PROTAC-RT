```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('R2.csv')
type=df['algorithm']
PActivity=df['R2']
```

```python
plt.style.use('ggplot')
sns.violinplot(x=type,
               y=df.R2,scale='width',bw=0.1,width=0.8,palette='Set3')
plt.savefig('R2.jpg',dpi=800)
plt.show()
```
